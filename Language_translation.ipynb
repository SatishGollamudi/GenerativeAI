```python
# Install dependencies first:
#!pip install transformers sentencepiece torch --upgrade

from transformers import MarianMTModel, MarianTokenizer

# Load English → Hindi model
en_hi_model_name = "Helsinki-NLP/opus-mt-en-hi"
en_hi_tokenizer = MarianTokenizer.from_pretrained(en_hi_model_name)
en_hi_model = MarianMTModel.from_pretrained(en_hi_model_name)
```

```python
# Load Hindi → English model
hi_en_model_name = "Helsinki-NLP/opus-mt-hi-en"
hi_en_tokenizer = MarianTokenizer.from_pretrained(hi_en_model_name)
hi_en_model = MarianMTModel.from_pretrained(hi_en_model_name)
```

```python
def translate_en_to_hi(text: str) -> str:
    """Translate English text to Hindi."""
    batch = en_hi_tokenizer([text], return_tensors="pt", padding=True)
    generated = en_hi_model.generate(**batch, max_new_tokens=128)
    return en_hi_tokenizer.decode(generated[0], skip_special_tokens=True)
```

```python
def translate_hi_to_en(text: str) -> str:
    """Translate Hindi text to English."""
    batch = hi_en_tokenizer([text], return_tensors="pt", padding=True)
    generated = hi_en_model.generate(**batch, max_new_tokens=128)
    return hi_en_tokenizer.decode(generated[0], skip_special_tokens=True)
```

```python
# ------------------ Demo ------------------
if __name__ == "__main__":
    print("=== English ↔ Hindi Translator ===")
    while True:
        print("\nChoose direction:")
        print("1. English → Hindi")
        print("2. Hindi → English")
        print("3. Exit")
        choice = input("Enter choice: ")

        if choice == "1":
            text = input("Enter English text: ")
            print("Hindi Translation:", translate_en_to_hi(text))
        elif choice == "2":
            text = input("Enter Hindi text: ")
            print("English Translation:", translate_hi_to_en(text))
        elif choice == "3":
            print("Exiting translator...")
            break
        else:
            print("Invalid choice, try again.")
```
